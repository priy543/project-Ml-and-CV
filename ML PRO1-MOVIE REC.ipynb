```python
import pandas as pd

movies = pd.read_csv(r"C:\Users\91739\Downloads\lab\Movie ML dataset\movies.csv")
ratings = pd.read_csv(r"C:\Users\91739\Downloads\lab\Movie ML dataset\ratings.csv")

print(movies.head())
print(ratings.head())

user_movie_matrix = ratings.pivot_table(index='userId', columns='movieId', values='rating')
print(user_movie_matrix.head())

from sklearn.metrics.pairwise import cosine_similarity

movie_similarity = cosine_similarity(user_movie_matrix.T.fillna(0))

import numpy as np

def recommend(movie_id, top_n=5):
    movie_index = user_movie_matrix.columns.tolist().index(movie_id)
    similarity_scores = list(enumerate(movie_similarity[movie_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    recommended_movies = [user_movie_matrix.columns[i[0]] for i in similarity_scores]
    return movies[movies['movieId'].isin(recommended_movies)][['title']]

recommend(1)
```
