```python
import pandas as pd

movies = pd.read_csv(r"C:\Users\91739\Downloads\lab\Movie ML dataset\movies.csv")
ratings = pd.read_csv(r"C:\Users\91739\Downloads\lab\Movie ML dataset\ratings.csv")
data = pd.merge(ratings, movies, on='movieId')
from sklearn.preprocessing import LabelEncoder

le_user = LabelEncoder()
le_movie = LabelEncoder()

data['user_enc'] = le_user.fit_transform(data['userId'])
data['movie_enc'] = le_movie.fit_transform(data['movieId'])
X = data[['user_enc','movie_enc']]
y = data['rating']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
from sklearn.linear_model import Ridge

model = Ridge(alpha=1.0)
model.fit(X_train, y_train)
from sklearn.metrics import mean_squared_error
pred = model.predict(X_test)

rmse = mean_squared_error(y_test, pred, squared=False)
print("RMSE:", rmse)
def ridge_recommend(user_id, top_n=5):
    user_encoded = le_user.transform([user_id])[0]
    movie_ids = data['movieId'].unique()
    movie_encoded = le_movie.transform(movie_ids)

    user_list = [user_encoded]*len(movie_ids)
    df = pd.DataFrame({'user_enc':user_list, 'movie_enc':movie_encoded})
    predictions = model.predict(df)

    rec_df = pd.DataFrame({'movieId':movie_ids,'pred_rating':predictions})
    rec_df = rec_df.merge(movies, on='movieId').sort_values(by='pred_rating', ascending=False)

    return rec_df[['title','pred_rating']].head(top_n)
ridge_recommend(1)

```
